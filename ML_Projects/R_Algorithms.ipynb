```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.svm import SVR
from sklearn.linear_model import BayesianRidge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv("CarPrice_dataset.csv")
```

```python
df.head()
```

```python

# separating the independent and dependent variables

# Independent variables - all columns except 'price'
X = df.drop(columns=['price'])  # Drop the 'price' column

# Dependent variable (target) - the 'price' column
y = df['price']

# Identify numeric columns
numeric_cols = X.select_dtypes(include=['number']).columns

# Filter to keep only numeric columns
X_numeric = X[numeric_cols]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_numeric, y, test_size=0.2, random_state=42)

# Scale the dataset
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# KNN Algorithm --- for regression 

k = 3  # Number of neighbors
model = KNeighborsRegressor(n_neighbors=k)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
r2 = r2_score(y_test, y_pred)
print(f"R² Score: {r2:.2f}")

# Print predictions
print("Predictions:  \n", y_pred)
```

```python
# SVM Algorithm -- for regression
# C -- regularization parameter, Epsilon -- epsilon-insensitive loss function used to measure the margin of tolerance where no errors are given in this margin.
# kernel = linear, creates linear line as a boundary to differentiate between the classes
svr_model = SVR(kernel='linear', C=100.0, epsilon=0.2) 
svr_model.fit(X_train, y_train)

# Make predictions
y_pred = svr_model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.2f}")
print(f"R² Score: {r2:.2f}")
```

```python
#Naive Bayes Algorithm -- for regression (using bayesian ridge regression since that is suitable for regression)

# Initialize Bayesian Ridge Regression
bayesian_ridge = BayesianRidge()
bayesian_ridge.fit(X_train, y_train)

# Make predictions
y_pred = bayesian_ridge.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.2f}")
print(f"R² Score: {r2:.2f}")
```

```python
# Decision Tree Algorithm -- for regression 

# Initialize and train the Decision Tree Regressor
regressor = DecisionTreeRegressor(random_state=42)
regressor.fit(X_train, y_train)

# Make predictions
y_pred = regressor.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.2f}")
print(f"R² Score: {r2:.2f}")
```

```python
# Random Forest Algorithm -- for regression 

# Initialize and train the Random Forest Regressor
forest_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
forest_regressor.fit(X_train, y_train)

# Make predictions
y_pred = forest_regressor.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.2f}")
print(f"R² Score: {r2:.2f}")
```
